## Problem 3D: Implementation of SGD

Fill in the loss, gradient, and SGD functions according to the guidelines given in the problem statement in order to perform SGD.

```python
def loss(X, Y, w):
    '''
    Calculate the squared loss function.
    
    Inputs:
        X: A (N, D) shaped numpy array containing the data points.
        Y: A (N, ) shaped numpy array containing the (float) labels of the data points.
        w: A (D, ) shaped numpy array containing the weight vector.
    
    Outputs:
        The loss evaluated with respect to X, Y, and w.
    '''
    return np.sum((Y - np.matmul(X,w))**2)

def gradient(x, y, w):
    '''
    Calculate the gradient of the loss function with respect to the weight vector w,
    evaluated at a single point (x, y) and weight vector w.
    
    Inputs:
        x: A (D, ) shaped numpy array containing a single data point.
        y: The float label for the data point.
        w: A (D, ) shaped numpy array containing the weight vector.
        
    Output:
        The gradient of the loss with respect to w. 
    '''

    return np.multiply(-2*(y-np.dot(w, x)),x,dtype=np.longdouble)

def SGD(X, Y, w_start, eta, N_epochs):
    '''
    Perform SGD using dataset (X, Y), initial weight vector w_start,
    learning rate eta, and N_epochs epochs.
    
    Inputs:
        X: A (N, D) shaped numpy array containing the data points.
        Y: A (N, ) shaped numpy array containing the (float) labels of the data points.
        w_start:  A (D, ) shaped numpy array containing the weight vector initialization.
        eta: The step size.
        N_epochs: The number of epochs (iterations) to run SGD.
        
    Outputs:
        W: A (N_epochs, D) shaped array containing the weight vectors from all iterations.
        losses: A (N_epochs, ) shaped array containing the losses from all iterations.
    '''
    
    N = np.shape(X)[0]
    D = np.shape(X)[1]
    
    W = np.zeros(shape=(N_epochs,D))
    
    losses = np.zeros(shape=(N_epochs, ))
    
    w = w_start
    
    for e in range(0,N_epochs):
        
        # Randomly go through points and update
        for i in np.random.permutation(N):
            w -= np.multiply(eta,gradient(X[i], Y[i], w),dtype=np.longdouble)
            
        W[e] = w
        losses[e] = loss(X, Y, w)
        print(w)
        
    return W, losses
```

```python
def loss(X, Y, w):
    '''
    Calculate the squared loss function.
    
    Inputs:
        X: A (N, D) shaped numpy array containing the data points.
        Y: A (N, ) shaped numpy array containing the (float) labels of the data points.
        w: A (D, ) shaped numpy array containing the weight vector.
    
    Outputs:
        The loss evaluated with respect to X, Y, and w.
    '''
    a=np.matmul(X,w)
    return np.sum(Y - np.matmul(a,a))

def gradient(x, y, w):
                  
    '''
    Calculate the gradient of the loss function with respect to the weight vector w,
    evaluated at a single point (x, y) and weight vector w.
    
    Inputs:
        x: A (D, ) shaped numpy array containing a single data point.
        y: The float label for the data point.
        w: A (D, ) shaped numpy array containing the weight vector.
        
    Output:
        The gradient of the loss with respect to w. 
    '''

    return -2*(y-np.dot(w, x))*x

def SGD(X, Y, w_start, eta, N_epochs):
    '''
    Perform SGD using dataset (X, Y), initial weight vector w_start,
    learning rate eta, and N_epochs epochs.
    
    Inputs:
        X: A (N, D) shaped numpy array containing the data points.
        Y: A (N, ) shaped numpy array containing the (float) labels of the data points.
        w_start:  A (D, ) shaped numpy array containing the weight vector initialization.
        eta: The step size.
        N_epochs: The number of epochs (iterations) to run SGD.
        
    Outputs:
        W: A (N_epochs, D) shaped array containing the weight vectors from all iterations.
        losses: A (N_epochs, ) shaped array containing the losses from all iterations.
    '''
    
    N = np.shape(X)[0]
    D = np.shape(X)[1]
    
    W = np.zeros(shape=(N_epochs,D),dtype=np.float64)
    
    losses = np.zeros(shape=(N_epochs, ), dtype=np.float64)
    
    w = w_start
    
    for e in range(0,N_epochs):
        # Randomly go through points and update
        for i in range(1,N):
            print(w)
            w -= eta*gradient(X[i], Y[i], w)
            
        W[e] = w
        losses[e] = loss(X, Y, w)
        
    return W, losses
```
